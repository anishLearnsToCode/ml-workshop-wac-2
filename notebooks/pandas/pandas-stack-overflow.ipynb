```python
# Percentage of people in each country who know python
country_population = survey['Country'].value_counts()
country_population
```

```python
python_users = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
python_users
```

```python
python_users_per_country = 100 * python_users / country_population
python_users_per_country
```

## Creating a New Table Using Concatenation

```python
python = pd.concat([country_population, python_users], axis='columns', sort=False)
python
```

```python
python.columns = ['population', 'python_users']
python
```

```python
python['percentage_python'] = 100 * python['python_users'] / python['population']
python
```

```python
# sorting on basis of python knowledge
python.sort_values(by='percentage_python', ascending=False)
```
